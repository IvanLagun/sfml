```python
df['LIMIT_R'] = pd.cut(df['LIMIT_BAL'], bins=[0, 10000, 100000, 200000, 400000, 700000, df.describe().loc['max', 'LIMIT_BAL']], labels=['A', 'B', 'C', 'D', 'E', 'F']).astype(object)
```

```python
df[df['LIMIT_R'] == 'F'][['LIMIT_R', 'LIMIT_BAL']].head()
```

```python
df_lit = pd.DataFrame (
    [
        ['A',[0, 10000]],
        ['B',[10001, 100000]],
        ['C',[100001, 200000]],
        ['D',[200001, 400000]],
        ['E',[400001, 700000]],
        ['F',[700001, df.describe().loc['max', 'LIMIT_BAL']]]
    ],
    columns=['LIT', 'LIT_LIMIT']
)
```

```python
def set_lit(limit):
    for lit in df_lit:
        if limit in lit[1]:
            return lit[0]
```
